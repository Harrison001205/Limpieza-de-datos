```python
import pandas as pd
import numpy as np
from google.colab import files

# 1. Creamos la base de datos (puedes saltar este paso si subes tu archivo)
datos_azar = {
    'VentaID': range(1, 1001),
    'Producto': np.random.choice(['Laptop', 'Mouse', 'Teclado', 'Monitor', 'mousE', None], 1000),
    'Precio': np.random.choice(['$1200', '$25', '$45', '$300', 'Error', None], 1000),
    'Fecha': np.random.choice(['2023-01-01', '01/02/2023', '2023-50-50', None], 1000),
    'Email': ['usuario@gmail.com' if i % 10 == 0 else f'user{i}@mail.com' for i in range(1000)]
}
tabla = pd.DataFrame(datos_azar)

# 2. Limpieza de la tabla
tabla['Producto'] = tabla['Producto'].str.strip().str.capitalize().fillna('No registrado')
tabla['Precio'] = tabla['Precio'].str.replace('$', '', regex=False)
tabla['Precio'] = pd.to_numeric(tabla['Precio'], errors='coerce')
tabla['Precio'] = tabla['Precio'].fillna(tabla['Precio'].mean())
tabla['Fecha'] = pd.to_datetime(tabla['Fecha'], errors='coerce')

# 3. Quitar errores y duplicados
tabla_limpia = tabla.dropna(subset=['Fecha']).drop_duplicates(subset=['Email'])

# 4. Guardar y descargar el resultado
tabla_limpia.to_csv('Resultado_Final.csv', index=False)
files.download('Resultado_Final.csv')

# 5. Ver el resultado en pantalla
tabla_limpia
```

```python
import matplotlib.pyplot as plt

# 1. CÁLCULOS Y CONCLUSIÓN
ingresos = tabla_limpia['Precio'].sum()
producto_top = tabla_limpia['Producto'].value_counts().idxmax()
total_filas = len(tabla_limpia)

print("--- REPORTE FINAL ---")
print(f"Total Ingresos: ${ingresos:,.2f}")
print(f"Producto más vendido: {producto_top}")
print(f"Registros procesados con éxito: {total_filas} (de 1,000)")
print("\nConclusión: La base de datos es confiable tras eliminar duplicados y errores de formato.")

# 2. GRÁFICA DE VENTAS
ventas_por_item = tabla_limpia.groupby('Producto')['Precio'].sum().sort_values()

plt.figure(figsize=(8, 4))
ventas_por_item.plot(kind='barh', color='skyblue')
plt.title('Ingresos por Producto')
plt.xlabel('Ventas ($)')
plt.tight_layout()
plt.show()
```
